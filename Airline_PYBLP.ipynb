```python
data.rename(columns={'Price':'prices'}, inplace=True)
data['prices']=data['prices']/1000
data.rename(columns={'share':'shares'}, inplace=True)

for i in range(11):
    data.rename(columns={'IV_'+str(i):'demand_instruments'+str(i-1)},inplace=True)
    
data.rename(columns={'com_dist':'demand_instruments10'}, inplace=True)
data.rename(columns={'aveincon':'demand_instruments11'}, inplace=True)

    


data['firm_ids'] = data.groupby('Down').ngroup()
data['up_firm_ids'] = data.groupby('Up').ngroup()
data['Nonstop']=(data.Connecting=="1")*1
data['market_ids']=data.MK+data.Q.astype(str)
data['clustering_ids']=data.market_ids
```

```python
dummies = pd.get_dummies(data['Up'],drop_first=True)

#op= pd.DataFrame(dummies)
fuel=np.transpose(np.array([data['fuel']]))
fuelop=np.multiply(dummies,fuel)

fuelop=pd.DataFrame(fuelop)
a=['demand_instruments'+str(i) for i in range(12,29)]

fuelop.columns=a
fuelop
data=pd.concat([data,fuelop],axis=1)
data
```

```python
# random coefficent nested logit (RCNL)
initial_sigma=np.diag([1,1,1,1,1])
df1['nesting_ids'] = 1
X1_formulation = pyblp.Formulation('0 + prices+Codeshare+incon+Nonstop+ns',absorb='C(Down)+C(Q)')
X2_formulation = pyblp.Formulation('1 + prices')
product_formulations = (X1_formulation, X2_formulation)
product_formulations

def solve_rcnl(df):
    groups = df.groupby(['market_ids', 'nesting_ids'])
    #df['demand_instruments20'] = groups['shares'].transform(np.size)
    X1_formulation = pyblp.Formulation('0 + prices+Codeshare+incon+Nonstop+ns',absorb='C(Down)+C(Q)')
    X2_formulation = pyblp.Formulation('1 + prices')
    rcnl_formulation = (X1_formulation, X2_formulation)
    mc_integration = pyblp.Integration('monte_carlo', size=50, specification_options={'seed': 0})
    problem = pyblp.Problem(rcnl_formulation, df,integration =mc_integration)
    bfgs = pyblp.Optimization('bfgs', {'gtol': 1e-4})
    return problem.solve(sigma=np.ones((2, 2)), optimization=bfgs,rho=0.4)
```

```python
rcnl_results1 = solve_rcnl(df1)
rcnl_results1
```

```python
# random coefficent nested logit (RCNL)
# full version
initial_sigma=np.diag([1,1,1,1,1])
df1['nesting_ids'] = 1
X1_formulation = pyblp.Formulation('0 + prices+Codeshare+incon+Nonstop+ns',absorb='C(Down)+C(Q)')
X2_formulation = pyblp.Formulation('1 + prices')
product_formulations = (X1_formulation, X2_formulation)
product_formulations

def solve_rcnl(df):
    groups = df.groupby(['market_ids', 'nesting_ids'])
    #df['demand_instruments20'] = groups['shares'].transform(np.size)
    X1_formulation = pyblp.Formulation('0 + prices+Codeshare+incon+Nonstop+ns',absorb='C(Down)+C(Q)')
    X2_formulation = pyblp.Formulation('0 + prices+Codeshare+incon+Nonstop+ns')
    rcnl_formulation = (X1_formulation, X2_formulation)
    mc_integration = pyblp.Integration('monte_carlo', size=500, specification_options={'seed': 0})
    problem = pyblp.Problem(rcnl_formulation, df,integration =mc_integration)
    bfgs = pyblp.Optimization('bfgs', {'gtol': 1e-14})
    return problem.solve(sigma=initial_sigma,W_type='clustered',se_type='clustered',initial_update=True, optimization=bfgs,rho=0.4)
rcnl_results2 = solve_rcnl(df1)
rcnl_results2
```

```python
RCNL_e=rcnl_results2.compute_elasticities()
diversion = rcnl_results2.compute_diversion_ratios()
```
